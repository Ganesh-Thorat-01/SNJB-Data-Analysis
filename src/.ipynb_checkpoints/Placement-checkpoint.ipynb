```python
def dash(value):
    value = str(value)
    if '-' in value:
        value = value.split('-')[0]
        return int(value)
    else:
        return int(value)
```

```python
placement_df["Academic Year"]=placement_df["Academic Year"].apply(dash)
```

```python
placement_df
```

```python
from fbprophet import Prophet
```

```python
df=placement_df[["Academic Year" 	,"Placement"]]
df
```

```python
# prepare expected column names
df.columns = ['ds', 'y']
#df['ds']= pd.to_datetime(df['ds'])
# define the model
model = Prophet()
# fit the model
model.fit(df)
```

```python
test=pd.DataFrame([2022,2020,2021,2020],columns=["ds"])
```

```python
test["ds"]=pd.to_datetime(df["ds"])
test
```

```python
model.predict(test)
```

```python
df
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
future = pd.DataFrame(df["ds"])
future.columns = ['ds']
```

```python
forecast = model.predict(future)
```

```python
y_true = df['y'].values
y_pred = forecast['yhat'].values
mae = mean_absolute_error(y_true, y_pred)
print('MAE: %.3f' % mae)
# plot expected vs actual
plt.plot(y_true, label='Actual')
plt.plot(y_pred, label='Predicted')
plt.legend()
plt.show()
```
